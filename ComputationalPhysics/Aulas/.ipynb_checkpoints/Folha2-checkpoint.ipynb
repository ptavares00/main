```python
#Al√≠nea 1
import numpy as np
import pylab as pl
r=np.array(list(np.loadtxt("sunspots.txt")),float)
pl.plot(r[:,0],r[:,1])
pl.ylabel("N¬∫ de Manchas Solares")
pl.xlabel("M√™s")
pl.show()
```

```python
#Al√≠nea 2
d=np.array(list(np.loadtxt("sunspots.txt")),float)
pl.plot(d[:,0],d[:,1])
pl.ylabel("N¬∫ de Manchas Solares")
pl.xlabel("M√™s")
pl.xlim(0,1000)
pl.show()
```

```python
#Al√≠near 3
l=list(d[0:5,1])
r=5
row,col=np.shape(d)
# descobrir n√∫mero de pontos: print(np.shape(d))
for i in range(5,row-5):
    soma=0
    for j in range(i-5,i+6):
        soma+=d[j,1]
    l.append(soma/(2*r+1))
for i in range(row-5,row):
    l.append(d[i,1])

d=np.array(list(np.loadtxt("sunspots.txt")),float)
pl.plot(d[:,0],d[:,1])
pl.plot(d[:,0],l)
pl.ylabel("N¬∫ de Manchas Solares / M√©dia m√≥vel")
pl.xlabel("M√™s")
pl.xlim(0,1000)
pl.show()
```
