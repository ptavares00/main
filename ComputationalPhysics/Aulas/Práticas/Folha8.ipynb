```python
M = 100
p = 1e-6
V = np.zeros((M + 1, M + 1), float)
V[M//5 : 4*M//5, M//5] = 1
V[M//5 : 4*M//5, 4*M//5] = -1
V_new = V.copy()

# é possível fazer melhor, utilizando o ciclo for na posição i e dividir em 3 ciclos -> diminui ~0.5s
def potencial():
    for j in range(1, M):
        V_new[1:-1, j] = 1/4 * (V[1:-1, j-1] + V[1:-1, j+1] + V[0:-2, j] + V[2:, j])
    V_new[M//5 : 4*M//5, M//5] = 1
    V_new[M//5 : 4*M//5, 4*M//5] = -1

potencial()
error = abs(np.max(V_new - V))

while error > p:
    V = V_new.copy()
    potencial()
    error = np.max(abs(V_new - V))

plt.imshow(V_new)
plt.colorbar()
plt.show()
```
