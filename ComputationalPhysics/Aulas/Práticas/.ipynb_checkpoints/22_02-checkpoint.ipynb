# Expansão em Série de Taylor

$$ exp(x)=1+x+\frac{x^2}{2!}+\frac{x^3}{3!}+... $$

```python
#menos eficaz
import math as mt
import numpy as np
import pylab as pl
%matplotlib inline 
x=[0.1, 0.3, 0.4, 0.6, 1.0, 1.5, 2.0, 4.0, 10, 20, 40]
N=40
taylor_exp=[]
xexp=[]
for x0 in x:
    exp_x0=1
    for k in range(1,N+1):
        exp_x0+=x0**k/mt.factorial(k)
    taylor_exp.append(exp_x0)
    xexp.append(mt.exp(x0))
pl.plot(x,xexp)
pl.plot(x,taylor_exp)
pl.yscale("log")
```

```python
#mais eficaz
x=[0.1, 0.3, 0.4, 0.6, 1.0, 1.5, 2.0, 4.0, 10, 20, 40]
N=40
taylor_exp=[]
xexp=[]
for i,x0 in enumerate(x):
    exp_x0=0
    a=1
    for k in range(N):
        exp_x0+=a
        a*=x0/(k+1)
    taylor_exp.append(exp_x0)
    xexp.append(mt.exp(x0))
pl.plot(x,xexp)
pl.plot(x,taylor_exp)
pl.yscale("log")
pl.show()
```

```python
#Com arrays
x=np.array([0.1, 0.3, 0.4, 0.6, 1.0, 1.5, 2.0, 4.0, 10, 20, 40],float)
taylor_exp=np.zeros(len(x))
an=np.ones(len(x))
for k in range(N):
    taylor_exp+=an
    an*=x/(k+1)
pl.plot(x,np.exp(x))
pl.plot(x,taylor_exp)
pl.yscale("log")
pl.show()
```

```python
x=np.linspace(0,35,100)
taylor_sin=np.zeros(len(x))
an=x.copy()
M=50
for k in range(M):
    taylor_sin+=an
    an*= - x**2/((2*k+3)*(2*k+2))
pl.plot(x,taylor_sin)
pl.plot(x,np.sin(x),'o')
pl.show()
```

# TPC
Expansão de taylor da função 1/(x+1)

```python
x=np.linspace(-0.9,1,40)
N=40
taylor_f=np.zeros(len(x))
a=np.ones(len(x))
for k in range(N):
    taylor_f+=a
    a*=(-1)*x
pl.plot(x, 1/(x+1), "c")
pl.plot(x, taylor_f, "o")
```
