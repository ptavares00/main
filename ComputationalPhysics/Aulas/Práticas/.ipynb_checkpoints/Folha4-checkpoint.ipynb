# Exercício 1

```python
import numpy as np
import matplotlib.pyplot as plt

#Função para calcular as distâncias
def trap_vel(t,v):
    h=1
    d=np.array([0],float)
    d1=h/2*(v[:-1]+v[1:])
    d=np.concatenate((d,d1))
    for i in range(1,len(d)):
        d[i]+=d[i-1]
    return d

#Importar dados e calcular as distâncias
dados=np.loadtxt("velocities.txt")
t,v=dados[:,0],dados[:,1]
d=trap_vel(t,v)

#Análise Gráfica
plt.plot(t,v,label="velocidade (m/s)")
plt.plot(t,d, label="distância (m)")
plt.xlabel("tempo (s)")
plt.legend()
```

# Exercício 2

```python
def f(x):
    return x**4-2*x+1

def simpson(a,b,N,f):
    h=(b-a)/N
    x=np.linspace(a,b,N+1)
    y=f(x)
    return h/3*np.sum(y[1::2]*4+y[2::2]+y[:-1:2])

def trapezio(a=0,b=2,N=1000,f=f):
    h=(b-a)/N
    x=np.linspace(a,b,N+1)
    y=f(x)
    return h/2*np.sum(y[1:]+y[:-1])

def ret(a=0,b=2,N=1000,f=f):
    h=(b-a)/N
    x=np.linspace(a,b,N+1)
    y=f(x)
    return h*np.sum(y[1:])

integral=simpson(0,2,10,f)
print(integral)
print("Erro(%) =", "{:.2f}%".format((integral-4.4)/4.4*100))


erro_T=np.empty(18)
erro_S=np.empty(18)
erro_R=np.empty(18)
x=[2**i for i in range(2,20)]
for k,ele in enumerate(x):
    erro_T[k]=trapezio(0,2,ele,f)
    erro_S[k]=simpson(0,2,ele,f)
    erro_R[k]=ret(0,2,ele,f)
erro_T=abs((erro_T-4.4*np.ones(18))/4.4*100)
erro_S=abs((erro_S-4.4*np.ones(18))/4.4*100)
erro_R=abs((erro_R-4.4*np.ones(18))/4.4*100)
        
    
plt.plot(x,erro_T,label="Erro Trapézio")
plt.plot(x,erro_S,label="Erro Simpson")
plt.plot(x,erro_R,label="Erro Retangular")
plt.legend()
plt.yscale("log")
plt.xscale("log")
print("Trapézio em %", erro_T)
print("Simpson em %", erro_S)
print("Retangular em %", erro_R)
```

# Exercício 7

```python
#Alínea a)

def trap_adap(p=1e-6,a=0,b=1): 
    
    f=lambda aux: np.sin((100*aux)**0.5)**2
    S = lambda N,h,f : h*np.sum(f(a+np.arange(1,N,2)*h))
    
    N=1
    I1=trapezio(a,b,N,f)
    
    erro=1
    
    while erro>p:
        N*=2
        h=(b-a)/N
        I2=I1/2+S(N,h,f)
        erro=abs(I2-I1)/3
        I1=I2
        
    return (N,I2,erro)

R=trap_adap()
print("N =", R[0])
print("I = " + "{0:.7f}".format(R[1]))
print("Erro = " + "{0:.7f}".format(R[2]))
```

```python
#Alínea b)

def romberg(p=1e-6,a=0,b=1):
    f = lambda x : np.sin((100*x)**0.5)**2
    S = lambda N,h,f : h*np.sum(f(a+np.arange(1,N,2)*h))
    
    R=np.zeros([50,50])
    i,N=0,100   #Escolher o N inicial
    R[0,0]=trapezio(a,b,N,f)
    erro=1
    
    while abs(erro) > p:
        i+=1
        N*=2
        h=(b-a)/N
        R[i,0]=R[i-1,0]/2+S(N,h,f)
        for m in range(i):
            erro=1/(4**(m+1)-1)*(R[i,m]-R[i-1,m])
            R[i,m+1]=R[i,m]+erro        
    return R[:i+1,:i+1]



M=romberg()
for k,r in enumerate(M):
    s="\t"
    for c in r[:k+1]:
        s+="{0:.7f}".format(c) + " | "
    print(s)
```

```python
#Verificar dependência do erro segundo o método de romberg
def romberg_2(i0,N,a=0,b=1):
    f = lambda x : np.sin((100*x)**0.5)**2
    S = lambda N,h,f : h*np.sum(f(a+np.arange(1,N,2)*h))
    
    R=np.zeros([50,50])
    i=0
    R[0,0]=trapezio(a,b,N,f)
    erro=1
    
    while i < i0:
        i+=1
        N*=2
        h=(b-a)/N
        R[i,0]=R[i-1,0]/2+S(N,h,f)
        for m in range(i):
            erro=1/(4**(m+1)-1)*(R[i,m]-R[i-1,m])
            R[i,m+1]=R[i,m]+erro        
    return erro
aux=[2**n for n in range(4,10)]
erro_l=np.zeros(6)
h_l=np.zeros(6)
for k in range(6):
    erro_l[k]=romberg(2,aux[k])   #Alterar o i0 para diferentes dependências do erro
    h_l[k]=1/aux[k]
plt.plot(h_l,erro_l,"c")
plt.plot(h_l,(h_l**4),"b")   #Verificar se o declive vai de acordo com o esperado
plt.xscale("log")
plt.yscale("log")
```

# Exercício 8

```python
def simpson_adap(p=1e-6,a=0,b=1):
    
    f = lambda x : np.sin((100*x)**0.5)**2
    T = lambda N,h,f : 2/3*np.sum(f(a+np.arange(1,N,2)*h))
    
    N=2   #N é par
    h=(b-a)/N
    S1=1/3*(f(a)+f(b)+2*np.sum(f(a+np.arange(2,N-1,2)*h)))
    T1=T(N,h,f)
    I1=simpson(a,b,N,f)
    erro=1
    
    while abs(erro)>p:
        N*=2
        h=(b-a)/N
        S2=S1+T1
        T2=T(N,h,f)
        I2=h*(S2+2*T2)
        erro=1/15*(I2-I1)
        S1,T1,I1=S2,T2,I2
        
    return (N,I2,erro)

S=simpson_adap(1e-6)
print("N =", S[0])
print("I = " + "{0:.7f}".format(S[1]))
print("Erro = " + "{0:.7f}".format(S[2]))
```

# Exercício 9

```python
def gaussxw(N):

    # Initial approximation to roots of the Legendre polynomial
    a = np.linspace(3,4*N-1,N)/(4*N+2)
    x = np.cos(np.pi*a+1/(8*N*N*np.tan(a)))

    # Find roots using Newton's method
    epsilon = 1e-15
    delta = 1.0
    while delta>epsilon:
        p0 = np.ones(N,float)
        p1 = np.copy(x)
        for k in range(1,N):
            p0,p1 = p1,((2*k+1)*x*p1-k*p0)/(k+1)
        dp = (N+1)*(p0-x*p1)/(1-x*x)
        dx = p1/dp
        x -= dx
        delta = max(abs(dx))

    # Calculate the weights
    w = 2*(N+1)*(N+1)/(N*N*(1-x*x)*dp*dp)

    return x,w

xw9=gaussxw(50)
def cv(T,xw=xw9):
    f=lambda x : x**4*np.exp(x)/(np.exp(x)-1)**2
    
    v=1e-3
    ro=6.022e28
    theta_d=428
    k_b=1.380649e-23
    a=0
    b=theta_d/T
    
    x,w=xw
    x=0.5*(b-a)*x+0.5*(b+a)
    w=0.5*(b-a)*w
    
    return 9*v*ro*k_b*b**(-3)*np.sum(w*f(x))

x=np.linspace(5,500,1000)
y=[cv(i) for i in x]
plt.figure(1,figsize=(5,5))
plt.plot(x,y)
plt.grid()
plt.xlim(0,500)
plt.ylim(0,2500)
plt.xlabel("Temperatura, T")
plt.ylabel("Capacidade Calorífica, cv")
plt.show()

#Descobrir o erro 
#N=20
#print("erro =", cv(50,gaussxw(2*N))-cv(50,gaussxw(N)))
```

# Exercício 10

```python
xw10=gaussxw(20)

def T(a,xw=xw10):
    f = lambda x : (a**4-x**4)**-0.5
    
    x,w=xw
    x=0.5*a*x+0.5*a
    w=0.5*a*w
    
    return 8**0.5*np.sum(w*f(x))

x=np.linspace(0,2,100)
y=[T(i) for i in x]
plt.figure(1,figsize=(5,5))
plt.plot(x,y)
plt.grid()
plt.xlabel("Amplitude, a")
plt.ylabel("Período, T")
plt.xlim(0,2.15)
plt.ylim(0,190)
plt.show()
```

# Exercício 11

```python
xw11=xw9

def I_I0(x, co, z, xw=xw11):
    f_C = lambda t: np.cos(np.pi/2*t**2)
    f_S = lambda t: np.sin(np.pi/2*t**2)
    
    u=x*(2/(co*z))**0.5
    xp,wp=xw
    xp=0.5*u*xp+0.5*u
    wp=0.5*u*wp
    
    C=np.sum(wp*f_C(xp))
    S=np.sum(wp*f_S(xp))
    
    return 1/8*((2*C+1)**2+(2*S+1)**2)

x=np.linspace(-5,5,500)
y=[I_I0(i,1,3) for i in x]
plt.figure(1,figsize=(5,5))
plt.plot(x,y)
plt.grid()
plt.xlabel("x (m)")
plt.ylabel("I / I0")
plt.show()
```

# 5.14 Livro

```python
xw14=gaussxw(200)

def Fz(z, xw=xw14):
    f = lambda x,y,z: (x**2+y**2+z**2)**-1.5
    
    N=200
    L=10
    G=6.674e-11
    sigma=10000/10**2
    a=-L/2
    b=L/2
    x0,w0=xw
    
    x=0.5*(b-a)*x0+0.5*(b+a)
    wi=0.5*(b-a)*w0
    y,wj=np.copy(x),np.copy(wi)
    
    z_=z*np.ones(N)
    I=np.sum([wi[k]*np.sum(wj*f(x[k]*np.ones(N),y,z_)) for k in range(N)])
    return I * G * z * sigma

z=np.linspace(0,10,100)
F=[Fz(ele) for ele in z]

plt.figure(1,figsize=(7,7))
plt.plot(z,F)
plt.grid()
plt.show()
```
