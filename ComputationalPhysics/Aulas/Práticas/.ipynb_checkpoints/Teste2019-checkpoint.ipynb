# 1.1

```python
import numpy as np
import scipy.linalg as sp
import matplotlib.pyplot as plt


# Exercício 1
L = 1; a = 1.95

def c2(x):
    return 1 + a * (x - L / 2)

def f(r, x, w):
    phi , psi = r
    f_phi = psi
    f_psi = - (w ** 2 / c2(x)) * phi
    return np.array([f_phi, f_psi], float)

def solution(w):
    n = 1000; h = L / n
    x_values = np.linspace(0, L, n+1)
    r = np.array([0, 1], float)
    r_values = np.empty([n+1, 2], float)
    for i, x in enumerate(x_values):
        r_values[i] = r
        k1 = h * f(r, x, w)
        k2 = h * f(r + k1/2, x + h/2, w)
        k3 = h * f(r + k2/2, x + h/2, w)
        k4 = h * f(r + k3, x + h, w)
        r += (k1 + 2 * k2 + 2 * k3 + k4) / 6
    return r_values[-1, 0]
```

# 1.2

Não consigo satisfazer sempre esta condição. Esta condição só é satisfeita para certas frequências w. O facto de a equação estar definida a menos de um fator multiplicativo implica que alterar as condições iniciais altera esta constante multiplicativa, mas através de alterar esta constante não conseguimos fazer com que a solução seja 0 em x = L.

O método de shooting consiste em escolher umas condições iniciais aleatórias (caso não se tenha intuição do valor que estas devem ter) e verificar se com estas as condições fronteira são estabelicidas. Usa-se um processo iterativo em que se variam as condições iniciais até que se encontre uma solução que satisfaça as condições fronteira. Utilizam-se métodos para procurar zeros de uma função, tal como o método da bijeção para fazer esta procura.

Contudo, neste problema o que interessa alterar são as frequências e não as condições iniciais, logo escolhe-se umas condições iniciais de modo a não ter a solução nula e altera-se a frequência até encontrar as frequências que satisfazem as condições fronteira, que neste caso é x = 0 (que é sempre satisfeita) e x = L (a que queremos satisfazer). Vou utilizar o método da bijeção (binary search) porque, embora tenham uma convergência assintótica exponencial (mais lento que o de Newton e da secante), considero que é adequado a este problema.

# 1.3

```python
def binary_search(x1, x2, f):
    p = 1e-6
    f1 = f(x1)
    f2 = f(x2)
    while abs(f1 - f2) > p:
        average = (x1 + x2) / 2
        f_average = f(average)
        if f_average * f1 > 0:
            x1 = average
            f1 = f_average
        else:
            x2 = average
            f2 = f_average
    return (x2 + x1) / 2

a = 1.95

x1 = 1; x2 = 3
w0 = binary_search(x1, x2, solution)
print(w0)

x1 = 3; x2 = 6
w1 = binary_search(x1, x2, solution)
print(w1)

x1 = 6; x2 = 9
w2 = binary_search(x1, x2, solution)
print(w2)

w_list = [w0, w1, w2]

def graph(w, r0):
    n = 5000; h = L / n
    x_values = np.linspace(0, L, n+1)
    r = r0.copy()
    r_values = np.empty([n+1, 2], float)
    for i, x in enumerate(x_values):
        r_values[i] = r.copy()
        k1 = h * f(r, x, w)
        k2 = h * f(r + k1/2, x + h/2, w)
        k3 = h * f(r + k2/2, x + h/2, w)
        k4 = h * f(r + k3, x + h, w)
        r += (k1 + 2 * k2 + 2 * k3 + k4) / 6
    return x_values, r_values

r0 = np.array([0, 1], float)
for i, w in enumerate(w_list):
    x_list, r_list = graph(w, r0)
    plt.plot(x_list, r_list[:, 0], label="n = {}".format(i))
plt.grid()
plt.legend()
plt.show()
```

# 2.1, 2.2 e 2.3

```python
n = 1000  # número de pontos
L = 1; dx = L / n
x_values = np.linspace(0, L, n + 1)
t_max = 50
h = 1e-2


# Matriz A
a = 0
A = np.identity((n + 1) * 2)
a1 = - h / 2 
for i in range(0, (n + 1) * 2 - 2, 2):
    A[i, i+1] = a1
    aux = a1 * c2(x_values[int(i // 2)]) / (dx ** 2)
    A[i+1, i+2] = aux
    A[i+1, i] = - aux * 2
    if i > 1:
        A[i+1, i-2] = aux
A[-2, -1] = a1
A[-1, -2] = - a1 * c2(x_values[int(-1)]) / (dx ** 2) * 2
A[-1, -4] = a1 * c2(x_values[int(-1)]) / (dx ** 2)

B = -A
np.fill_diagonal(B, 1)
lu, piv = sp.lu_factor(A)

def crank_nicolson():
    t_values = np.arange(0, t_max, h)
    y = np.zeros((n + 1) * 2)
    y[0:-1:2] = np.sin(np.pi * x_values)
    y_values = np.empty((len(t_values), (n + 1) * 2), float)
    
    for i, t in enumerate(t_values):
        y_values[i] = y
        y = sp.lu_solve((lu, piv), B @ y)
    return t_values, y_values

t_list, y = crank_nicolson()
plt.plot(t_list, y[:, n])
plt.show()
```

# 2.4

```python
psi = y[:, n]
coeficientes = np.fft.rfft(psi)

w_max =  2 * np.pi / h
w_list = np.linspace(0, w_max, len(t_list))
print(len(w_list))
print(len(coeficientes))


plt.figure(figsize=(16,8))
plt.plot(w_list[:2501], abs(coeficientes))
plt.xlim(0, 30)
plt.show()
```
