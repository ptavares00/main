```python
import numpy as np
import matplotlib.pyplot as plt
import cmath as cm

# Transformada de Fourier Discreta
def dft(y):
    N = len(y)
    coefficients = np.empty(N, complex)
    for k in range(N):
        coefficients[k] = np.sum([y[n] * cm.exp(-2j * np.pi * k * n / N) for n in range(N)])
    return coefficients


def f(n, N): return cm.sin(np.pi * n / N) * cm.sin(20 * np.pi * n / N)


f = np.vectorize(f)
y_ret = np.ones(1000)
y_dente = np.arange(1000)
y_sin = [f(n, 1000) for n in range(1000)]

coefficients_ret = dft(y_ret)
coefficients_dente = dft(y_dente)
coefficients_sin = dft(y_sin)

plt.Figure(figsize=(8, 8))
plt.plot(abs(coefficients_ret))
plt.show()

plt.Figure(figsize=(8, 8))
plt.plot(abs(coefficients_dente))
plt.show()

plt.Figure(figsize=(8, 8))
plt.plot(abs(coefficients_sin))
plt.show()
```

```python
# Com o numpy

coefficients_ret_np = np.fft.fft(y_ret)
coefficients_dente_np = np.fft.fft(y_dente)
coefficients_sin_np = np.fft.fft(y_sin)

plt.Figure(figsize=(8, 8))
plt.plot(abs(coefficients_ret))
plt.show()

plt.Figure(figsize=(8, 8))
plt.plot(abs(coefficients_dente))
plt.show()

plt.Figure(figsize=(8, 8))
plt.plot(abs(coefficients_sin))
plt.show()
```

```python
# Transformada de Fourier Inversa
def idft(c):
    N = len(c)
    y = np.empty(N, complex)
    for n in range(N):
        y[n] = 1 / N * np.sum([c[k] * cm.exp(2j * np.pi * k * n / N) for k in range(N)])
    return y


plt.plot(idft(coefficients_ret))
plt.ylim(0,2)
plt.show()

plt.plot(idft(coefficients_dente))
plt.show()

plt.plot(idft(coefficients_sin))
plt.show()
```
