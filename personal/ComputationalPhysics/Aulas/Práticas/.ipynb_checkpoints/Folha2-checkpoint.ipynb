```python
#Al√≠near 3 sem sum function
d=np.array(list(np.loadtxt("sunspots.txt")),float)
l=list(d[0:5,1])
r=5
row,col=np.shape(d)
# descobrir n√∫mero de pontos: print(np.shape(d))
for i in range(5,row-5):
    soma=0
    for j in range(i-5,i+6):
        soma+=d[j,1]
    l.append(soma/(2*r+1))
for i in range(row-5,row):
    l.append(d[i,1])


pl.plot(d[0:1000,0],d[0:1000,1])
pl.plot(d[0:1000,0],l[0:1000])
pl.ylabel("N¬∫ de Manchas Solares / M√©dia m√≥vel")
pl.xlabel("M√™s")
pl.show()
```

```python
#Al√≠near 3 com sum function
d=np.array(list(np.loadtxt("sunspots.txt")),float)
l=list(d[0:5,1])
r=5
row,col=np.shape(d)
# descobrir n√∫mero de pontos: print(np.shape(d))
for i in range(5,row-5):
    l.append(np.sum(d[i-5:i+6,1])/(2*r+1))
for i in range(row-5,row):
    l.append(d[i,1])


pl.plot(d[0:1000,0],d[0:1000,1])
pl.plot(d[0:1000,0],l[0:1000])
pl.ylabel("N¬∫ de Manchas Solares / M√©dia m√≥vel")
pl.xlabel("M√™s")
pl.show()
```

# Exerc√≠cio 5

```python
import vpython as vp
import math as mt
import numpy as np

#Par√¢metros
c1=2500
cs=750
c2=0.5

#Dados planetas
data=np.array([[2440,57.9e6,88.0],
               [6052,108.2e6,224.7],
               [6371,149.6e6,365.3],
               [3386,227.9e6,687.0],
               [69173,778.5e6,4331.6],
               [57316,1433.4e6,10759.2]],float)
cores=[vp.color.white,vp.color.orange,vp.color.blue,vp.color.magenta,vp.color.gray(0.5),vp.color.yellow]
data[:,0]*=c1
#Frequ√™ncia Angular
data[:,2]=(2*mt.pi)/data[:,2]

#Representar planetas
sol=vp.sphere(radius=695500*cs,pos=vp.vector(0,0,0),color=vp.color.yellow)
planetas=np.empty(len(data),vp.sphere)
for i in range(len(data)):
    planetas[i]=vp.sphere(radius=data[i,0],pos=vp.vector(data[i,1]+695500*cs,0,0),color=cores[i])
    
#Anima√ß√£o
t=0
while True:
    vp.rate(30)
    for j in range(len(planetas)):
        planetas[j].pos=vp.vector((data[j,1]+695500*cs)*mt.cos(data[j,2]*t),(data[j,1]+695500*cs)*mt.sin(data[j,2]*t),0)
    t+=1*c2
```

# Exerc√≠cio 6

```python
import math as mt
import numpy as np
import pylab as pl
x0=1/2
r=1
def mapa_xvsx_(x0,r):
    x=[]
    x_=[]
    i=0
    while i!=1000:
        x.append(x0)
        x0=r*x0*(1-x0)
        x_.append(x0)
        i+=1
    pl.scatter(x,x_, s=10, color="black")
    pl.xlim(0,0.60)
    pl.ylim(0,0.30)
    pl.xlabel("x", size=20, color="blue")
    pl.ylabel("x'", size=20, color="blue")
    pl.grid()
    pl.show()

mapa_xvsx_(x0,r)
```

```python
def mapa_rvsx(x0,r):
    x_=[]
    i=0
    while i!=1000:
        x0=r*x0*(1-x0)
        x_.append(x0)
        i+=1
    r_=np.ones(len(x_))*r
    pl.scatter(r_,x_,s=1,color="black")

for r in np.arange(1,4,0.01):
    mapa_rvsx(0.5,r)
    
pl.xlabel("r", size=20, color="blue")
pl.ylabel("x", size=20, color="blue")
pl.grid()
pl.show()
```

```python
#N√£o sei qual o gr√°fico que √© suposto aparecer (o de cima ou o de baixo)
#Mais Eficaz
r=np.arange(1,4,0.0001)
x=np.ones(len(r))*0.5
i=0
while i!=1000:
    x=r*x*(1-x)    
    i+=1
pl.scatter(r,x,s=1,color="black")
pl.xlabel("r", size=20, color="blue")
pl.ylabel("x", size=20, color="blue")
pl.grid()
pl.show()
```

# Exerc√≠cio 7

```python
def Mandelbrot(c,n=100):
    z=0+0j
    z_mod=0
    i=0
    while i != n and z_mod<=2:
        z=z**2+c
        z_mod=(z.real**2+z.imag**2)**0.5
        i+=1
    return i==n
```

```python
N=1000
grelha=np.empty([N,N],float)
espa√ßamento=4/N
for i in range(N):
    x=-2+espa√ßamento*i
    for k in range(N):
        y=-2+espa√ßamento*k
        grelha[i,k]=Mandelbrot(complex(x,y))
pl.imshow(grelha, origin="lower", extent=[-2,2,-2,2])
pl.gray()
pl.colorbar()
pl.xlabel("Parte Real", size=15)
pl.ylabel("Parte Imagin√°ria", size=15)
```

```python
def Mandelbrot2(c,n=100):
    z=0+0j
    z_mod=0
    i=0
    while i != n and z_mod<=2:
        z=z**2+c
        z_mod=(z.real**2+z.imag**2)**0.5
        i+=1
    return i

N=1000
grelha=np.empty([N,N],float)
espa√ßamento=4/N
for i in range(N):
    x=-2+espa√ßamento*i
    for k in range(N):
        y=-2+espa√ßamento*k
        grelha[i,k]=Mandelbrot2(complex(x,y))
pl.imshow(grelha, origin="lower", extent=[-2,2,-2,2])
pl.viridis()
pl.colorbar()
pl.xlabel("Parte Real")
pl.ylabel("Parte Imagin√°ria")
pl.clim(0,100)
```
