```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def hist(amostra, a, b, M):
    dx = (b - a) / M
    N = len(amostra)
    h = np.zeros(M)
    indices = ((amostra - a) / (b - a) * M).astype(int)
    for i in indices:
        h[i] += 1
    x = np.linspace(a + dx / 2, b - dx / 2, M)
    return x, h / dx / N


x, h0 = hist(np.random.random_sample(128), 0, 1, 128)
x, h1 = hist(np.random.random_sample(1024), 0, 1, 128)
x, h2 = hist(np.random.random_sample(4096), 0, 1, 128)
plt.figure(figsize=(16,10))
plt.plot(x, h0, label="128")
plt.plot(x, h1, label="1024")
plt.plot(x, h2, label="4096")
plt.legend()
plt.show()

med0 = np.zeros(128)
for i in range(500):
    x, h0 = hist(np.random.random_sample(4096), 0, 1, 128)
    med0 += (h0 - med0) / (i + 1)
    
plt.figure(figsize=(16,10))
plt.plot(x, med0)
plt.show()
```
